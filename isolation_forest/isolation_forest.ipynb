```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
# load
df = pd.read_csv('~/Documents/data-scientist-toolkit/datasets/carros.csv', sep=',') 
```

```python
df.info()
```

```python
output = np.zeros((3, df.size//2))
```

```python
c = [0.01, 0.05, 0.1]
for i in range(len(c)):
    model = IsolationForest(contamination = c[i]).fit(df)
    output[i] = model.predict(df)
```

```python
plt.set_cmap('jet')
fig = plt.figure(figsize=(13, 4))

for i in range(len(c)):
    ax = fig.add_subplot(1, 3, i+1)
    ax.scatter(df[output[i]==-1].iloc[:, 0],
            df[output[i]==-1].iloc[:, 1],
            c='skyblue', marker='s', s=500)

    ax.scatter(df.iloc[:, 0],
            df.iloc[:, 1],
            c=range(df.size//2), marker='x',
            s=500, alpha=0.6)

    ax.set_title('Contamination: %0.2f' % c[i], 
                size=18, color='purple')
    ax.set_ylabel('Price ($)', size=10)
    ax.set_xlabel('Kms Utilized', size=14)

plt.show()
```
